## Lambda

```python
import pandas as pd
```

```python
# creating dataframe
data={
    'name':['Sujan Katwal','Dammar Khatri','Abiraj katwal '],
    'email':['iamsuzank@gmail.com','kc.dammar@hotmail.com','abi@yahoo.com']
    
}
```

```python
dataset=pd.DataFrame(data)
dataset.to_csv('dataset.csv')#saving to csv file
```

```python
dataset
```

```python
# display only first name
first_name=lambda x:x.split(' ')[0]
dataset['first name']=dataset.name.apply(first_name)
```

```python
display(dataset)
```

```python
# get name after @
domain=lambda x:x.split('@')[-1]
dataset['domain']=dataset['email'].apply(domain)
dataset
```

```python
dataset['hourly_wage']=500
dataset['hours_worked']=[10,12,14]
```

```python
dataset
```

```python
total_earned=lambda row:(row.hourly_wage*10)+((row.hours_worked-10)*(1.5*row.hourly_wage)) \
if row.hours_worked>10 \
else (row.hours_worked*row.hourly_wage)
dataset['total_earned']=dataset.apply(total_earned,axis=1)
dataset['Overtime_amount']=dataset.apply(lambda row:(row.hours_worked-10)*(1.5*row.hourly_wage) \
                                         if row.hours_worked>10 \
                                         else 0,axis=1)

```

```python
dataset
```

### Changing Columns Name

```python
df = pd.DataFrame({
    'name': ['John', 'Jane', 'Sue', 'Fred'],
    'age': [23, 29, 21, 18]
})
display(df)
df.columns = ['First Name', 'Age']
display(df)
```

```python
df = pd.DataFrame({
    'name': ['John', 'Jane', 'Sue', 'Fred'],
    'age': [23, 29, 21, 18]
})
display(df)
df.rename(columns={
    'name': 'First Name'},
    inplace=True)
# renaming only one columns
display(df)
df.rename(columns={
    'First Name': 'First2 Name',
    'age': 'Age'},
    inplace=True)
# Renaming multiple columns
display(df)
```

```python
import pandas as pd

orders = pd.read_csv('shoefly.csv')
display(orders.head())

# vegan if shoe material is not leather else animal
shoe_source=lambda x: 'vegan' if (x.shoe_material!='leather') else \
'animal'

# Dear Mr.last_name if gender is male else Dear Ms.
orders['shoe_source']=orders.apply(shoe_source,axis=1)
salutation=lambda x: ('Dear Mr. ' + x.last_name) if (x.gender=='male') else\
('Dear Ms. ' + x.last_name)

orders['salutation']=orders.apply(salutation,axis=1)
display(orders.head())
```
